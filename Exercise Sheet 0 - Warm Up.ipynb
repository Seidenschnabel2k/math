```python
fig, ax = plt.subplots()
x = rainfall
y = umba_sold
n = len(x)
m = (len(x)*(np.sum(x*y)) - np.sum(x)*np.sum(y))/(len(x)*sum(x**2)-np.sum(x)**2)
b = (np.sum(x**2)*np.sum(y) - np.sum(x)*np.sum(x*y))/(len(x)*sum(x**2)-np.sum(x)**2)
print(f"m = {m}\nb = {b}")
def f(x):
    return m*x+b


ax.plot(rainfall,label='rainfall')
ax.plot(umba_sold,label="Umbrl")
ax.plot(f(rainfall),'m',label="predict")
ax.set_xlabel("Month")
ax.legend()
```

```python
a = np.array([2,3])
b = np.array([-1,4])

def a3(x,y):
    e = y/np.linalg.norm(y)
    #e = y/(y[0]**2+y[1]**2)**.5
    para = (x@e)*e
    return para, x-para
    

para, orto= a3(a,b)
plt.quiver(0, 0, a[0],a[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, b[0],b[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, para[0],para[1], angles='xy', scale_units='xy', scale=1, color='g')
plt.quiver(0, 0, orto[0],orto[1], angles='xy', scale_units='xy', scale=1, color='m')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-2, 5)
plt.ylim(-1, 4)
```

```python
mean = -3
sigma = 2
N = 1000
s = np.random.normal(mean, sigma, N)

def f(x):
    return 1/(2*np.pi)**.5/sigma * np.exp(-((x-mean)**2)/(2*sigma**2))
    
#dataPnts = sigma * np.random.randn(N) + mean
#plt.hist(dataPnts,bins=50)
print(np.mean(s),np.std(s))
fig, ax = plt.subplots(figsize=(6,4))

ax.scatter(range(N),s)

fig, ax = plt.subplots(figsize=(6,4))
ax.hist(s, bins=50, density=True, alpha=0.6, color='darkgrey',label="Histogram")
plotvalues = np.linspace(mean-4*sigma,mean+4*sigma,1000)
ax.plot(plotvalues,f(plotvalues),'k',label="Normal Dist")
ax.plot([mean,mean], [0,f(mean)],'k--', label="Average")
ax.plot([mean+sigma,mean+sigma], [0,f(mean+sigma)],'r:', label='1$\\sigma$ Intervall')
ax.plot([mean-sigma,mean-sigma], [0,f(mean-sigma)],'r:')
ax.plot([mean+sigma*2,mean+sigma*2], [0,f(mean+sigma*2)],'g:', label='2$\\sigma$ Intervall')
ax.plot([mean-sigma*2,mean-sigma*2], [0,f(mean-sigma*2)],'g:')
ax.legend()
plt.show()
```

```python
range_s1 = s[s>(mean-sigma)]
range_s1 = range_s1[range_s1<(mean+sigma)]
print(len(range_s1)/N)
range_s2 = s[s>(mean-sigma*2)]
range_s2 = range_s2[range_s2<(mean+sigma*2)]
print(len(range_s2)/N)
range_s3 = s[s>(mean-sigma*3)]
range_s3 = range_s3[range_s3<(mean+sigma*3)]
print(len(range_s3)/N)
```
